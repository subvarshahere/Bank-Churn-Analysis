```python
#are old customers less active?
df=pd.read_csv("test.csv")
sns.lineplot(x='Age',y='IsActiveMember',data=df)
plt.show()
```

```python
#balance vs churn
dt_data=df[(df['IsActiveMember']==1) & df['Balance']>0]
sns.boxplot(x='IsActiveMember',y='Balance', data=dt_data)
plt.show()
```

```python
#Tenure vs Churn
tenure_churn=df[(df['Tenure']>=5) & (df['IsActiveMember']==1)]
sns.boxplot(x='IsActiveMember', y='Tenure', data=tenure_churn)
plt.show()
```
